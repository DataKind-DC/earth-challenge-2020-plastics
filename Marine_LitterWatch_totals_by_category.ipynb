```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
df2 = pd.read_csv('MarineLitterWatch_data.csv')
```

```python
df2['Plastic Raw Count'] = df2.loc[:,'G1':'G124'].sum(axis=1)
df2['Rubber Raw Count'] = df2.loc[:,'G125':'G134'].sum(axis=1)
df2['Cloth Raw Count'] = df2.loc[:,'G137':'G145'].sum(axis=1)
df2['Processed Lumbar Raw Count'] = df2.loc[:,'G147':'G172'].sum(axis=1)
df2['Metal Raw Count'] = df2.loc[:,'G174':'G199'].sum(axis=1)
df2['Glass Raw Count'] = df2.loc[:,'G200':'G210'].sum(axis=1)
df2['Other Raw Count'] = df2.loc[:,'G211':'G213'].sum(axis=1)
df2['Fishing Gear Raw Count'] = 0
df2['Dataset'] = 'Marine LitterWatch'
df2.index.name = 'UniqueID'
df2['Year'] = df2['EventDate'].astype(str).str[:4]
df2['Month'] = df2['EventDate'].astype(str).str[4:6]
df2['Day'] = df2['EventDate'].astype(str).str[6:]
df2['Total Area Unit'] = 'meters'
df2['Total Item Count'] = df2.loc[:,'Plastic Raw Count':'Other Raw Count'].sum(axis=1)
df2
```

```python
df2 = df2[['Dataset','EventDate','Year','Month','Day','lat_y1','lon_x1','lat_y2','lon_x2','BeachCountrycode',
           'BeachLength_m','Total Area Unit','Total Item Count','Plastic Raw Count','Cloth Raw Count','Rubber Raw Count',
          'Processed Lumbar Raw Count','Metal Raw Count','Glass Raw Count','Other Raw Count','Fishing Gear Raw Count']]
df2.columns = ['Dataset','Date','Year','Month','Day','Latitude 1','Longitude 1','Latitude 2','Longitude 2','Country',
           'Total Area','Total Area Unit','Total Item Count','Plastic Raw Count','Cloth Raw Count','Rubber Raw Count',
          'Processed Lumbar Raw Count','Metal Raw Count','Glass Raw Count','Other Raw Count','Fishing Gear Raw Count']
df2
```

```python
df2.to_csv('Marine_LitterWatch_processed_updated.csv')
```
